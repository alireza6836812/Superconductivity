```python
import re
from IPython.display import clear_output

def calculator2(my_dict,string):
    def separate(string):
        my_list12 = []
        pattern = r'([A-Z][a-z]*)(\d*\.?\d*)'
        element = re.findall(pattern, string)
        for i in element:
            i = list(i)
            for j in i:
                if j == '':
                    i.pop()
                    i.append('1')
            my_list12.append(tuple(i))
        return(my_list12)

    total = 0
    output = 0

    for i in separate(string):
        element = i[0]
        try:
            my_dict[element]
        except:
            my_dict[element] = 0
        try:
            number = float(i[1])
            total += number
        except:
            pass
        output1 = my_dict[element]*number

        output = output + output1     
    output = output/total
    #clear_output()
    return(output)
```
