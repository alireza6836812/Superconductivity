```python
compounds = pd.read_csv('Cif_data_with_id_for_compounds.csv').drop(['Unnamed: 0'],axis=1)
compounds['reduced formula'] = ''
compounds
```

```python
k = 0
for i in compounds['compound']:
    try:
        if compounds['type'].iloc[k] == 'CuO':
            my_list12 = []
            pattern = r'([A-Z][a-z]*)(\d*\.?\d*)'
            element = re.findall(pattern, i)
            for ii in element:
                if ii[0] == 'Cu':
                    x = ii[1]
                if ii[0] == 'O':
                    y = ii[1]
                if ii[1] == '':
                    ii = (ii[0],'1')
                ii = list(ii)
                ii[1] = str(float(ii[1]) * 100)            
                for j in ii:
                    if j == '':
                        ii.pop()
                        ii.append('1')
                my_list12.append(tuple(ii))
            string = ''
            number_list = []
            for z in my_list12:
                number_list.append(int(float(z[1])))
            try:
                if (float(y)/float(x)) < 2:
                    j = compounds[((compounds.formula == i))].index
                    compounds = compounds.drop(j)
                else:
                    pass
            except:
                pass
        k += 1
    except:
        pass
compounds
```

```python
k = 0
for i in compounds['compound']:
        my_list12 = []
        pattern = r'([A-Z][a-z]*)(\d*\.?\d*)'
        element = re.findall(pattern, i)
        for i in element:
            if i[1] == '':
                i = (i[0],'1')
            i = list(i)
            i[1] = str(float(i[1]) * 10000)            
            for j in i:
                if j == '':
                    i.pop()
                    i.append('1')
            my_list12.append(tuple(i))
        string = ''
        number_list = []
        for l in my_list12:
            number_list.append(int(float(l[1])))
        a = np.gcd.reduce(number_list)    
        for l in my_list12:            
            string += (l[0] + str(float(l[1])/a)[-3::-1][::-1])
        compounds['reduced formula'].iloc[k] = string
        k += 1
clear_output()
compounds
```
